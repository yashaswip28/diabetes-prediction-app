```python
selected_features = variables
print(selected_features)
# data = pd.get_dummies(data, columns=['gender', 'smoking_history'], drop_first=True)
data = data[selected_features + ['diabetes']]
data
```

```python
data.head()
```

```python
bmi=data[['bmi']]
bmi.head()
```

```python
data['gender'].replace({'Female': 0, 'Male': 1, 'Other': 2}, inplace=True)
data['smoking_history'].replace({
    'No Info': 0, 'current': 1, 'ever': 2,
    'former': 3, 'never': 4, 'not current': 5
}, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
X = data.drop('diabetes', axis=1)
y = data['diabetes']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
print(X.columns.tolist())

```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import joblib

```

```python

X_train, X_test, y_train, y_test = tts(X_scaled_df, y, test_size=0.2, random_state=42)

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

```

```python

#we never have to use linear regression in these type of cases where the diabetes result in the data has only two states of having 
# not having 1 and 0 respectively as linear regression is used for numerical calculations involving price prediction.

# we hence use logistic regression where it uses logical function to map the values to 0s or 1s.
```

```python
# predicating 
y_pred_log = log_reg.predict(X_test)
```

```python
# Confusion Metrix
metrics.confusion_matrix(y_test,y_pred_log )
```

```python
acc_log = metrics.accuracy_score(y_test, y_pred_log)
acc_log
```

```python
print(metrics.classification_report(y_test, y_pred_log))
```

```python
#joblib.dump(model, 'logistic_model.pkl')
```

```python
# verifying / validating the efficiency using random forest
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
```

```python
print("Confusion Matrix - Random Forest:\n", metrics.confusion_matrix(y_test, y_pred_rf))
```

```python
from sklearn.metrics import accuracy_score

acc_log = accuracy_score(y_test, y_pred_log)
acc_rf = accuracy_score(y_test, y_pred_rf)

print("Logistic Regression Accuracy:", acc_log*100 , '%')
print("Random Forest Accuracy:", acc_rf*100 , '%')
```

```python
# validating the accuracy using decision trees.

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
# Create and fit the decision tree classifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)

# Perform predictions on the test set
y_pred = model.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test,y_pred)

print("Accuracy: ", accuracy*100 , '%')
```
